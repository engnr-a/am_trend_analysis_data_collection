# Followed Companies



```python

names_to_remove = ['IBM', 'Microsoft', 'Google', 'Harvard Business Review', 'Medtronic', 'The Wall Street Journal', 'Forbes',
                   'Meta', 'Ferrari', 'The Economist', 'Amazon Web Services (AWS)','The Economist', 'Tesla', 'OpenAI',
                   'Generative AI', 'SpaceX', 'Google Cloud', 'Mercedes-Benz AG', 'Shopzu.com', 'Porsche AG', 'Google for Developers',
                   'Boston', 'Dynamics', 'Google DeepMind', 'AdamDanyal.com', 'MIT Technology Review', 'Porsche AG',
                    "DeepLearning.AI","NVIDIA AI","Towards Data Science","Data Science Academy","Boston Dynamics","Microsoft Research","Machine Learning Mastery","Towards AI",
                    "AWS Machine Learning","Analytics Vidhya","UBS Asset Management","Real Python","PyTorch",
                   "BIG - Bjarke Ingels Group","Neuralink","OpenCV","Machine Learning","Volkswagen","Maven Analytics",
    "Advanced MS EXCEL, Power Query, Power BI skills for Business Intelligence","storytelling with data","Data Visualization Society",
    "Open Data Science Conference (ODSC)","BMW","SuperDataScience", "Mercedes-Benz Deutschland", "Audi Deutschland", "Czech Technical University in Prague",
                   "The Boring Company", "Porsche Deutschland GmbH", "Papers with Code", "Icon","High5", "Hamburg University of Technology"
                   
                    ]

df = pd.read_csv('./data/followed_companies_info.csv')
# Remove unwanted company names
df_filtered = df[~df['Company Name'].isin(names_to_remove)]

# Remove all rows where the URL column valye is "No URL"
df_filtered = df_filtered[df_filtered['URL'] != 'No URL']

# Reindex t
df_filtered = df_filtered.reset_index(drop=True)
df_filtered['Index'] = df_filtered.index + 1

# Remove commas, remove ' followers' text, and convert to integer...for exploratory analysis
df_filtered['Followers'] = df_filtered['Followers'].str.replace(',', '') 
df_filtered['Followers'] = df_filtered['Followers'].str.replace(' followers', '').astype(int)

#df_filtered
```

```python

df_sorted = df_filtered.sort_values(by='Followers', ascending=False)

top_n = 50
df_top_companies = df_sorted.head(top_n)
df_top_companies 
```

```python
df_sorted = df_filtered.sort_values(by='Followers', ascending=False)

top_n = 10
df_top_companies = df_sorted.head(top_n)

# Plotting
plt.figure(figsize=(8, 6))  
sns.barplot(x='Followers', y='Company Name', data=df_top_companies, palette="viridis")
plt.title(f'Top {top_n} Companies by Followers')
plt.xlabel('Number of Followers')
plt.ylabel('Company Name')
plt.show()
```

```python

df = pd.read_csv('./data/3Dadept_2024-04-07_09-53-07posts.csv')
df
```

```python
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['Date'])

df_grouped = df.groupby(df['date'].dt.to_period("M")).size().reset_index(name='counts')

df_grouped['date'] = df_grouped['date'].dt.to_timestamp()

df_grouped = df_grouped.sort_values(by='date', ascending=False)

# Plotting
plt.figure(figsize=(10, 6))  
sns.barplot(x='date', y='counts', data=df_grouped, palette="coolwarm")
plt.xticks(rotation=45)  
plt.title('Monthly Counts from Latest to Past')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from nltk.corpus import stopwords
import nltk


nltk.download('stopwords')

#df = 
all_words = ' '.join(df['post text'].str.lower()).split()


stop_words = set(stopwords.words('english'))
filtered_words = [word for word in all_words if word not in stop_words and word.isalpha()]  # isalpha() removes punctuation


word_counts = Counter(filtered_words)


top_20_words = word_counts.most_common(20)


df_top_words = pd.DataFrame(top_20_words, columns=['Word', 'Count'])

plt.figure(figsize=(10, 8))
sns.barplot(x='Count', y='Word', data=df_top_words, palette="magma")
plt.title('Top 20 Most Common Words')
plt.xlabel('Count')
plt.ylabel('Word')
plt.show()
```
