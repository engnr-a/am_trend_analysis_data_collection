# Tweets Collection By Date

```python
import pandas as pd
import glob

#data_folder = "data/keyphrase_search_results_raw/by_date/01_2024"
data_folder = "data/keyphrase_search_results_raw/by_date"

csv_files = glob.glob(f"{data_folder}/**/*.csv", recursive=True)

df = pd.concat((pd.read_csv(file) for file in csv_files), ignore_index=True)

df['Date'] = pd.to_datetime(df['Date'])
df
```

### Review Duplicate Tweets

```python
duplicates = df[df.duplicated(subset=['Date', 'Author ID'], keep=False)]

print("Duplicates based on 'Date' and 'Author ID':")
duplicates

duplicate_count = duplicates.shape[0]
print(f"Number of duplicate rows: {duplicate_count}")
```

### Review Tweet By Latest and Earliest

```python
sorted_df = df.sort_values(by='Date', ascending=True)
sorted_df
```

```python
earliest_date = df['Date'].min()
latest_date = df['Date'].max()

print(f"Earliest Date: {earliest_date}")
print(f"Latest Date: {latest_date}")
```

## Create Global Record of Unique Tweets

```python

df['Date_str'] = df['Date'].dt.strftime('%Y-%m-%dT%H:%M:%S.%f').str[:-3] + 'Z'

df['tweet_unique_key'] = df['Date_str'] + '_' + df['Author ID']

deduplicated_df = df.drop_duplicates(subset=['tweet_unique_key'])

deduplicated_df.reset_index(drop=True, inplace=True)
deduplicated_df['Index'] = deduplicated_df.index + 1

deduplicated_df.sort_values(by='Date', inplace=True)

detailed_filename = "unique_tweet_records.csv"
deduplicated_df[['Index', 'Date', 'tweet_unique_key']].to_csv(detailed_filename, index=False)

minimal_filename = "unique_keys.csv"
deduplicated_df[['tweet_unique_key']].to_csv(minimal_filename, header=['tweet_unique_key'], index=False)

print(f"Detailed file saved as: {detailed_filename}")
print(f"Minimal file saved as: {minimal_filename}")
```
