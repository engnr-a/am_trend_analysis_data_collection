```python
import pandas as pd
import glob


data_folder = "data/hash_additive_manufacturing"

csv_files = glob.glob(f"{data_folder}/*.csv")

combined_df = pd.concat((pd.read_csv(file) for file in csv_files), ignore_index=True)
combined_df['Date'] = pd.to_datetime(combined_df['Date'])
combined_df
```

```python
# Find duplicate rows based on 'Date' and 'Author ID'
duplicate_rows = combined_df[combined_df.duplicated(subset=['Date', 'Author ID'], keep=False)]
duplicate_rows.sort_values(by=['Date', 'Author ID'])
```

```python
unique_rows = combined_df.drop_duplicates(subset=['Date', 'Author ID'])

# Get the count of unique rows
unique_count = unique_rows.shape[0]

print(f"Number of unique rows based on 'Date' and 'Author ID': {unique_count}")
```

```python
# Ensure the 'Date' column is in datetime format
combined_df['Date'] = pd.to_datetime(combined_df['Date'])

# Drop duplicates based on 'Date' and 'Author ID'
unique_rows = combined_df.drop_duplicates(subset=['Date', 'Author ID'])

# Extract the month from the 'Date' column
unique_rows['Month'] = unique_rows['Date'].dt.to_period('M')

# Count posts by month
post_count_by_month = unique_rows.groupby('Month')['Post Text'].count()

# Print the count
print(post_count_by_month)
```
